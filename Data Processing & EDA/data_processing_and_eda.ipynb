```python
import re

# Function to extract specifications
def clean_specifications(specs):
    processor = re.search(r'(Intel|AMD|Apple).*Processor.*\)', specs)
    ram = re.search(r'(\d+\s?GB\s(?:DDR[34]|LPDDR[34]|Unified))', specs)
    storage = re.search(r'(\d+\s?GB|\d+\s?TB)\s?(SSD|HDD|eMMC|NVMe)', specs)
    os = re.search(r'(Windows \d+(?: Pro| Home)?|MacOS(?: [\w\s.]+)?|Linux(?: [\w\s.]+)?|Chrome OS(?: [\w\s.]+)?|DOS)', specs, re.IGNORECASE)
    display = re.search(r'(\d+\.?\d*\s?cm(?:\s?\(\d+\s?Inch\))?)\s?Display', specs, re.IGNORECASE)
    warranty = re.search(r'(\d+\sYear)', specs)

    return {
        "Processor": processor.group(0) if processor else None,
        "RAM": ram.group(0) if ram else None,
        "Storage": storage.group(0) if storage else None,
        "Operating System": os.group(0) if os else None,
        "Display": display.group(0) if display else None,
        "Warranty": warranty.group(0) if warranty else None,
    }
```

```python
specs_df = df['SPECIFICATIONS'].apply(lambda x: clean_specifications(str(x)))
df['clean_specs'] = specs_df
specs_df = pd.json_normalize(specs_df)

# Merge back to the original dataframe
df = pd.concat([df, specs_df], axis=1)
```

```python
# Function to clean and convert price to integer
def clean_price(price):
    # Remove non-digit characters (₹, commas, spaces)
    cleaned_price = re.sub(r'[^\d]', '', str(price))
    
    # Convert to integer if valid
    return int(cleaned_price) if cleaned_price.isdigit() else None

# Apply the cleaning function
df['PRICE'] = df['PRICE'].apply(clean_price)
```
