```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
import pandas as pd
import time
import os
```

```python
# Initilazing driver
options = Options()
options.add_argument('--headless')  # Run in headless mode
options.add_argument('--disable-gpu')
options.add_argument('--no-sandbox')
options.add_argument('--disable-blink-features=AutomationControlled')

driver = webdriver.Chrome(options=options)
driver.maximize_window()

# Get the desired Url
driver.get('https://www.flipkart.com/')
time.sleep(5)
```

```python
# No Sign In or login
try:
    wait = WebDriverWait(driver, 10)
    no_login = wait.until(EC.presence_of_element_located((By.CLASS_NAME, '_30XB9F')))
    ActionChains(driver).move_to_element(no_login).click().perform()
except:
    pass

# Clear Search bar and search for 'Laptos'
search = driver.find_element(By.NAME, 'q')
search.clear()
search.send_keys('Laptops')
search.send_keys(Keys.ENTER)
assert 'No results found' not in driver.page_source
```

```python
# Wait until page loads
wait = WebDriverWait(driver, 100)
wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'hEjLuS')))

# Creating lists to store the relevant Data
name_list = []
price_list = []
rating_list = []
specs_list = []
link_list = []
```

```python
# Loop through multiple pages (e.g., up to 92 pages)
i = 0
while i < 92:
    print(f'Scraping Web Page {i+1:.<25}')

    # All product containers with title
    products = driver.find_elements(By.CLASS_NAME, 'cPHDOP')
    print(f'Found {len(products)} products on this page.')

    for product in products:
        try:
            name = product.find_element(By.CLASS_NAME, 'KzDlHZ').text
        except:
            name = ''
        try:
            price = product.find_element(By.CLASS_NAME, '_4b5DiR').text
        except:
            price = ''
        try:
            rating = product.find_element(By.CLASS_NAME, 'XQDdHH').text
        except:
            rating = ''
        try:
            specs = product.find_element(By.CLASS_NAME, '_6NESgJ').text
        except:
            specs = ''
        try:
            link = product.find_element(By.TAG_NAME, 'a').get_attribute('href')
        except:
            link = ''

        name_list.append(name)
        price_list.append(price)
        rating_list.append(rating)
        specs_list.append(specs)
        link_list.append(link)

    # Move to next page
    try:
      driver.implicitly_wait(2)
      next_page = driver.find_element(By.LINK_TEXT, 'NEXT')
      next_page.click()
      time.sleep(2)
    except:
      print("No more pages or 'NEXT' button missing.")
      break

    i += 1
```

```python
# Store data into dictionary
info_dict = {
    'LAPTOP NAME': name_list,
    'PRICE': price_list,
    'RATING': rating_list,
    'SPECIFICATIONS': specs_list,
    'PRODUCT URL': link_list
}

# Save to DataFrame
info_df = pd.DataFrame(info_dict)
info_df.head()
```
